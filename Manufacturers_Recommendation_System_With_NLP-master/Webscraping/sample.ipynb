```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://www.thomasnet.com/local.html?cov=NA&heading=45340403&what=Machining&which=prod&radius=100&zip=73071&searchsource=searchwithin_radius&sort=&navsec=modify'
with urllib.request.urlopen(url) as response:
 html = response.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
list_of_company_url = []

tags = soup.find_all (class_ ="profile-card__title")
for tag in tags:
    list_of_company_url.append(tag.a.attrs['href'])    
```

```python
# list_of_company_url
```

```python
with urllib.request.urlopen('https://www.thomasnet.com'+ list_of_company_url[0]) as response:
    html1 = response.read()
soup1 = BeautifulSoup(html1, 'html.parser')
```

```python
Company_Name= soup1.find(class_="codetail").find('h1').get_text()
Phone_Number= soup1.find(class_="phoneline").span.get_text()
Email= soup1.find(class_="primary").get('href') 
company_description= soup1.find('div', id= 'copro_about').p.get_text()
Company_address= soup1.find(class_="addrline").get_text()[:-6]
Capabilities = soup1.find(class_="submenu").find_all('a')[1].get('href')

with urllib.request.urlopen('https://www.thomasnet.com'+ Capabilities) as response:
    Capabilities_Url = response.read()
soup2 = BeautifulSoup(Capabilities_Url, 'html.parser')
```

```python
soup2
# Capabilities_list= soup2.find(class_="alphablock M").find(class_="alphagroup pscols").find(class_="expand show").find_all('li')
Capabilities_list= soup2.find(class_="card", id="allprodserv").find_all('li')
```

```python
Manufacturer_Machining_capabilities = [] 
for element in Capabilities_list:  
    Manufacturer_Machining_capabilities.append(element.a.get_text())   
```
