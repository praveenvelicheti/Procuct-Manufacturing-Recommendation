```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
file1 = open("Links","r+") 
```

```python
df = pd.read_csv("Links",header=None)[0]
Links_of_Manufactureres = df.values.tolist()
```

```python
with urllib.request.urlopen('https://www.thomasnet.com'+ Links_of_Manufactureres[0]) as response:
    html1 = response.read()
soup = BeautifulSoup(html1, 'html.parser')
```

```python
Company_Name= soup.find(class_="codetail").find('h1').get_text()
Phone_Number= soup.find(class_="phoneline").span.get_text()
Email= soup.find(class_="primary").get('href') 
company_description= soup.find('div', id= 'copro_about').p.get_text()
Company_address= soup.find(class_="addrline").get_text()[:-6]
Capabilities_link = soup.find(class_="submenu").find_all('a')[1].get('href')

with urllib.request.urlopen('https://www.thomasnet.com'+ Capabilities_link) as response:
    Capabilities_data = response.read()
Company_Capabilities_data = BeautifulSoup(Capabilities_data, 'html.parser')
```

```python

# Capabilities_list= soup2.find(class_="alphablock M").find(class_="alphagroup pscols").find(class_="expand show").find_all('li')
Capabilities_list= Company_Capabilities_data.find(class_="card", id="allprodserv").find_all('li')
Capabilities_machine_list =Company_Capabilities_data.find(class_="alphablock M").find(class_="alphagroup pscols").find_all(class_="expand show")
zzzz =[]
for i in Capabilities_machine_list:
    print(i.a.get_text())
    print("######################################################")
    if (i.a.get_text()== 'Machining'):
        
        assd = i.find_all('li')
        for x in assd:  
            zzzz.append(x.a.get_text())
        print(((zzzz)))
        
```
